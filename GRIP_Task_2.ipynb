##**4. Determining the optimal value of K (number of clusters)**

```python
df_cluster = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
df_cluster.head()
```

###**4.1 Analyzing Value of WCSS for different values of 'K' in the range 1 to 9**

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df_cluster)
    sse.append(km.inertia_)

print("Value of Centroid (WCSS) for K ranging between 1 - 9 : ")
sse
```

###**4.2 Visualizing Elbow Plot**

```python
plt.figure(figsize=(9,7))
plt.grid()
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within cluster sum of squares (WCSS)')
plt.title("Elbow Plot")
plt.scatter(k_rng, sse ,  color = "blue" , label = "Centroids" )
plt.plot(k_rng , sse  , color="red" , label = "Elbow Plot")

#plt.plot(range(1,11), within_cluster_sum_of_squares, label = "Elbow graph", color = "#0000DD") 
```

###**4.3 Observations:**

* From the above elbow plot it is evident that, optimal value of k = 3
* Therefore, number of clusters = 3

##**5. Implementing K-means Classifier**

```python
km = KMeans(n_clusters=3)       #Default Parameters remain unchanged
km
```

###**5.1 Predicting cluster for each records**

```python
y_kmeans = km.fit_predict(df_cluster)
y_kmeans
```

###**5.2 Adding predicted values of cluster to the dataset**

```python
df_cluster["cluster"] = y_kmeans
df_cluster.head()
```

```python
df_cluster = np.array(df_cluster)    #coverting dataframe to numpy array for data manipulation
```

##**7. Evaluation of K-means Classifier using Silhouette Score**

* It depicts the consistency within clusters of data

```python
from sklearn.metrics.cluster import silhouette_score

result = silhouette_score(df_cluster, y_kmeans, metric = 'euclidean')
print("Silhouette Score: %.2f"%result)
```

###**7.1 Observations**

* Silhouette Score is close to the optimum value 1, which indicates good distinction of all the clusters

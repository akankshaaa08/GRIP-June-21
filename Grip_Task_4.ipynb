#**2. Reading Images**

```python
# creating figure
fig = plt.figure(figsize=(30,10))
  
# setting values to rows and column variables
rows = 2
columns = 4
  
# storing the image path
Image1 = cv2.imread('/content/drive/MyDrive/Images/book.jpg')
Image2 = cv2.imread('/content/drive/MyDrive/Images/car.jpg')
Image3 = cv2.imread('/content/drive/MyDrive/Images/metro_bridge_banner-.jpg')
Image4 = cv2.imread('/content/drive/MyDrive/Images/form.jpg')
Image5 = cv2.imread('/content/drive/MyDrive/Images/invoice.jpg')
Image6 = cv2.imread('/content/drive/MyDrive/Images/youtube-caption.png')
Image7 = cv2.imread('/content/drive/MyDrive/Images/quote.jpg')
Image8 = cv2.imread('/content/drive/MyDrive/Images/receipt.jpg')
  

# Adding a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
  
# showing image
plt.imshow(cv2.cvtColor(Image1, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Book")

#--------------------------------------------------------------------------------------


# Adding a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(cv2.cvtColor(Image2, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Number Plate")


#--------------------------------------------------------------------------------------
  
# Adding a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(cv2.cvtColor(Image3, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Banner")


#--------------------------------------------------------------------------------------
  
# Adding a subplot at the 4th position
fig.add_subplot(rows, columns, 4)
  
# showing image
plt.imshow(cv2.cvtColor(Image4, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Application Form")


#--------------------------------------------------------------------------------------

# Adding a subplot at the 5th position
fig.add_subplot(rows, columns, 5)
  
# showing image
plt.imshow(cv2.cvtColor(Image5, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Invoice")


#--------------------------------------------------------------------------------------
  
# Adding a subplot at the 6th position
fig.add_subplot(rows, columns, 6)
  
# showing image
plt.imshow(cv2.cvtColor(Image6, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Youtube Caption")


#--------------------------------------------------------------------------------------
  
# Adding a subplot at the 7th position
fig.add_subplot(rows, columns, 7)
  
# showing image
plt.imshow(cv2.cvtColor(Image7, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Quote")


#--------------------------------------------------------------------------------------

# Adding a subplot at the 8th position
fig.add_subplot(rows, columns, 8)
  
# showing image
plt.imshow(cv2.cvtColor(Image8, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Receipt")
```

#**3. Reading Characters**

```python
reader = easyocr.Reader(['en'] , gpu = True)

result_1 = reader.readtext(Image1)
result_2 = reader.readtext(Image2)
result_3 = reader.readtext(Image3)
result_4 = reader.readtext(Image4)
result_5 = reader.readtext(Image5)
result_6 = reader.readtext(Image6)
result_7 = reader.readtext(Image7)
result_8 = reader.readtext(Image8)

print("\nCharacter Detection Successful!")
```

#**4. Visualization**

##**4.1 Type of Image: Receipt**

```python
for read in result_8:
  print("\n-->",read[1],", Accuracy: ",read[2])
```

```python
for detection in result_8: 
    top_left = tuple([int(val) for val in detection[0][0]])
    bottom_right = tuple([int(val) for val in detection[0][2]])
    text = detection[1]
    font = cv2.FONT_HERSHEY_SIMPLEX
    Image8 = cv2.rectangle(Image8 ,top_left,bottom_right,(0,255,0),3)
    Image8 = cv2.putText(Image8 ,text,top_left, font, 1,(0,255,0),2,cv2.LINE_AA)


plt.figure(figsize=(10,10))
plt.imshow(Image8)
plt.show()
```
